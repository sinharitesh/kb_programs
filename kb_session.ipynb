## Git Setup

```python
import subprocess
from pathlib import Path

repo = '/app/data/kb_programs'
token = Path('/app/data/.gh_token').read_text().strip()

# Set remote URL with token
subprocess.run(['git', 'remote', 'set-url', 'origin',
    f'https://{token}@github.com/sinharitesh/kb_programs.git'], cwd=repo)
print('Remote configured')
```

## Git Push

```python
from subprocess import run

r = run(['git', '-C', repo, 'add', '-A'], capture_output=True, text=True)
print('ADD:', r.returncode)

r = run(['git', '-C', repo, 'commit', '-m', 'your commit message here'], capture_output=True, text=True)
print('COMMIT:', r.stdout.strip())

r = run(['git', '-C', repo, 'push'], capture_output=True, text=True)
print('PUSH:', r.stdout.strip(), r.stderr.strip())
```
